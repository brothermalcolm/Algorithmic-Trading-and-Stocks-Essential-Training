## Test algorithm accuracy

```python
# excel 03_03
df_0303 = pd.read_excel('./03_03_Begin.xlsx', skip_footer=3)
df_0303.tail()
```

```python
from matplotlib import style
style.use('ggplot')
df_0303.set_index(['Date']).iloc[:,1:6].plot.line()
```

```python
df_0303['Stand Dev.'] = df_0303['Adj Close'].rolling(100).std().shift(1)
df_0303['Big Buy'] = df_0303['Adj Close'] + df_0303['Stand Dev.'] < df_0303['Moving Avg']
df_0303['Big Buy Ret.'] = (df_0303['Big Buy'].shift(1) * df_0303['1 Day Return']).astype('float')
df_0303[98:].head(10)
```

## Algorithm profitability and trading decisions
e.g.

```python
df_0304 = df_0303
df_0304.tail(10)
```

```python
df_0304['1 Day Long Only'] = df_0304['Buy'] * df_0304['1 Day Return']
df_0304['1 Day Short Only'] = (df_0304['Buy'] == False) * -df_0304['1 Day Return']
df_0304['Big Buy Cap Used Ret'] = df_0304['Big Buy Ret.'].apply(lambda x: x if x != 0 else None) 
df_0304.tail(10)
```
