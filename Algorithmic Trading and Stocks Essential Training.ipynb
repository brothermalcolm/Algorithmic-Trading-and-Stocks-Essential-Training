```python
# excel 03_02

import pandas as pd

df_0302 = pd.read_excel('./03_02_Begin.xlsx', skip_footer=6)
df_0302.tail()
```

```python
df_0302.info()
```

```python
df_0302['Buy Column'] = df_0302['VIX'] < df_0302['Moving Avg. VIX'] 
df_0302['L-S Mean Reversion'] = df_0302['Buy Column'] * df_0302['Long Vol Ret'] 
df_0302.head()
```
